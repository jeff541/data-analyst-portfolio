# dataset https://www.kaggle.com/datasets/aiaiaidavid/the-big-dataset-of-ultra-marathon-running?resource=download

Objectifs: Analyse des données de plus de 1,3 million de marathons pour identifier les facteurs clés permettant de gagner une course.

Consignes: 

- Néttoyer et transformer les données
- Ne traiter que les courses de 50km ou de 50 miles qui ont eu lieu aux USA en 2020 
- Différence de vitesse entre les hommes et les femmes sur les courses de 50 km et 50 miles.
- Quels groupes d'âge performent le mieux dans la course de 50 miles (avec un minimum de 20 courses) ?
- Quels groupes d'âge ont les moins bonnes performances dans la course de 50 miles (avec un minimum de 10 courses) ?
- Vitesse moyenne des participants en fonction de la saison
- Vitesse moyenne des participants en fonction de la saison uniquement pour les courses de 50 miles

```python
import pandas as pd     #importons les bibliothèques
import seaborn as  sns
```

```python
df = pd.read_csv("TWO_CENTURIES_OF_UM_RACES.csv") # chargeons le fichier dans un dataframe
```

```python
# nétoyage des données
```

```python
#Nous voulons uniquement des courses aux États-Unis, 50 km ou 50 miles, 2020
```

```python
df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020)] # première étape trouver les courses de  50K ou 50miles qui ont eu lieu 2020
```

```python
df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA'
]  #trouver les courses aux USA
```

```python
#Combiner les requêtes
df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA') ]
```

```python
df2 = df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA') ] #création d"un autre dataframe df2 qui aurra uniquement les données qui nous intéressent
```

```python
df2.head(10) #controle des 10 premières lignes
```

```python
df2.shape #(nombre de lignes, nombre de colonnes )
```

```python
#Rétirons (USA) du noms des évènements 
df2['Event name']= df2['Event name'].str.split('(').str.get(0)
```

```python
df.head() #controle des 10 premières lignes
```

```python
# Ajout de la colonne athlete_age qui trouvera l'age actuelle des athlète. le code devra être adapté en fonction de l'année en cours
df2['Athlete_age'] = 2025 - df2['Athlete year of birth']
```

```python
df2.head() #controle des 10 premières lignes
```

```python
# retirer h dans dans les enregistrements de la colonne athlete performance
df2['Athlete performance']=df2['Athlete performance'].str.split('h').str.get(0)
```

```python
df2.head() #controle des 10 premières lignes
```

```python
# supprimer les colonnes: Athlete club , Athlete country, Athlete year of birth, Athlete Age Category
df2 = df2.drop(['Athlete club','Athlete country', 'Athlete year of birth','Athlete age category'], axis = 1 )
```

```python
df2.head() #controle des 10 premières lignes
```

```python

df2.isna().sum() # compter les valeurs nulles pour chacune des colonnes 
```

```python
df2[df2['Athlete_age'].isna() == 1]
#dans la colonne athlete_age récuperer les lignes qui ont une valeur nulle
```

```python
df2 = df2.dropna() #supprimer les lignes qui on une valeur nulle
```

```python
df2.shape #(nombre le lignes, nombre de colonne)
```

```python
#rechercher les doublons
```

```python
df2[df2.duplicated() == True]
```

```python
# nouvel index
df2.reset_index(drop = True )
```

```python
#attribuons les bons types
```

```python
df2.dtypes
```

```python
df2['Athlete_age'] = df2['Athlete_age'].astype(int) # convertir en entier
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float) # convertir en float
```

```python
df2.head() #controle des 10 premières lignes
```

```python
df2.dtypes
```

```python
# renommer les colonnes
df2 = df2.rename(columns = {
'Year of event': 'year',
'Event dates':'race_day',
'Event name':'race_name',
'Event distance/length':'race_length',        
'Event number of finishers':'race_number_of_finishers',
'Athlete performance':'athlete_performance',
'Athlete gender':'athlete_gender',
'Athlete average speed':'athlete_average_speed',
'Athlete ID':'athlete_id',
'Athlete_age':'athlete_age' 
})
```

```python
df2.head() #controle des 10 premières lignes
```

```python
#reordonner les colonnes dans un nouveau dataframe df3
```

```python
df3 = df2[['race_day','race_name','race_length','race_number_of_finishers','athlete_id','athlete_gender','athlete_average_speed','athlete_id','athlete_age']]
```

```python
#Réponse aux questions
```

```python
#Différence de vitesse entre les hommes et les femmes sur les courses de 50 km et 50 miles.
```

```python
df3.groupby(['race_length','athlete_gender'])['athlete_average_speed'].mean()
```

```python
#Quels groupes d'âge performent le mieux dans la course de 50 miles (avec un minimum de 20 courses) ?
```

```python
df3.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = False).query('count>19')
```

```python
#Quels groupes d'âge ont les moins bonnes performances dans la course de 50 miles (avec un minimum de 10 courses)
```

```python
df3.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = True).query('count>9')
```

```python
#Les saisons sont définies comme suit :

#Printemps : Mars (3) - Mai (5)
#Été : Juin (6) - Août (8)
#Automne : Septembre (9) - Novembre (11)
#Hiver : Décembre (12) - Février (2)
```

```python
df3['race_month']=df3['race_day'].str.split('.').str.get(1).astype(int) #créer une colonne moiuis à partir de la date
```

```python
df3.head() #controle des 10 premières lignes
```

```python
df3['race_season']=df3['race_month'].apply(lambda x: 'Winter' if x > 11 else  'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if  x > 2 else 'Winter') #créer une colonne saison et la remplir en fonction du mois de l'année
```

```python
df3.head() #controle des 10 premières lignes
```

```python
df3.groupby(['race_season'])['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = False) # vitesse moyenne des participants en fonction de la saison
```

```python
# vitesse moyenne des participants en fonction de la saison uniquement pour les courses de 50 miles
```

```python
df3.query('race_length=="50mi"').groupby(['race_season'])['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = False)
```
